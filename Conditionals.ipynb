<div class="alert alert-block alert-success">
<b>Determine if a triangle is equilateral, isosceles, or scalene</b>
</div>

```python
side1 = float(input("Enter side 1: "))
side2 = float(input("Enter side 2: "))
side3 = float(input("Enter side 3: "))

if (side1 + side2 > side3) and (side1 + side3 > side2) and (side1 + side3 > side2):
    if (side1 == side2) and (side1 == side3):
        print("The given dimensions are of an Equilateral triangle")
    elif (side1 == side2) or (side1 == side3) or (side2 == side3):
        print("The given dimensions are of an Isosceles triangle")
    else:
        print("The given dimensions are of a scalene triangle")
else:
    print("The given dimensions are not of a triangle")
```

<div class="alert alert-block alert-success">
<b>Calculate electricity bill based on units consumed (tiered pricing)</b>
</div>

```python
units = float(input("Enter units consumed: "))

if units <= 100:
    bill = units * 3.50
elif units <= 200:
    bill = (100 * 3.50) + ((units - 100) * 4.50)
elif units <= 300:
    bill = (100 * 3.50) + (100 * 4.50) + ((units - 200) * 5.50)
else:
    bill = (100 * 3.50) + (100 * 4.50) + (100 * 5.50) + ((units - 300) * 6.50)

print(f"Units: {units} = Bill: Rs.{bill:.2f}")
```

<div class="alert alert-block alert-success">
<b>Determine the quadrant of a coordinate point (x, y)</b>
</div>

```python
x = float(input("Enter x coordinate: "))
y = float(input("Enter y coordinate: "))

if x > 0 and y > 0:
    print(f"({x},{y}) is in the 1st quadrant")
elif x > 0 > y:
    print(f"({x},{y}) is in the 4th quadrant")
elif y > 0 > x:
    print(f"({x},{y}) is in the 2nd quadrant")
elif x < 0 and y < 0:
    print(f"({x},{y}) is in the 3rd quadrant")
elif x == 0 and y == 0:
    print(f"({x},{y}) lies on the origin")
elif x == 0 :
    print(f"({x},{y}) lies on the Y-axis")
else:
    print(f"({x},{y}) lies on the X-axis")
```

<div class="alert alert-block alert-success">
<b>Check if three numbers can form a Pythagorean triplet</b>
</div>

```python
a = int(input("Enter the 1st number: "))
b = int(input("Enter the 2nd number: "))
c = int(input("Enter the 3rd number: "))
# I used the integer conversion and not the float conversion, because to be a Pythagorean triplet
# all the numbers must be integers.

if (pow(a, 2) + pow(b, 2)  == pow(c, 2)) or (pow(a, 2) + pow(c, 2) == pow(b, 2)) or (pow(b, 2) + pow(c, 2) == pow(a, 2)):
    print(f"{a} and {b} and {c} are Pythagorean triplets")
else:
    print(f"{a} and {b} and {c} are not Pythagorean triplets")
```

<div class="alert alert-block alert-success">
<b>Create simple calculator using if-elif-else</b>
</div>

```python
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
operation = input ("Select operation: + , - , * , / ")

if operation == "+":
    print(f"{num1} + {num2} = {num1 + num2}")
elif operation == "-":
    print(f"{num1} - {num2} = {num1 - num2}")
elif operation == "*":
    print(f"{num1} * {num2} = {num1 * num2}")
elif operation == "/":
    print(f"{num1} / {num2} = {num1 / num2}")
else:
    print("Error: Invalid operation")
```

<div class="alert alert-block alert-success">
<b>Check if a number is a perfect square</b>
</div>

```python
import math

num = int(input("Enter a number: "))

if num < 0:
    print(f"{num} is not a perfect square (negative numbers cannot be perfect squares)")
else:
    root = math.isqrt(num)

    if root * root == num:
        print(f"{num} is a perfect square ({root} * {root} = {num})")
    else:
        print(f"{num} is not a perfect square")
```
