<div class="alert alert-block alert-success">
<b>Find the average of 5 numbers</b>
</div>

```python
num1 = float(input("Enter number 1: "))
num2 = float(input("Enter number 2: "))
num3 = float(input("Enter number 3: "))
num4 = float(input("Enter number 4: "))
num5 = float(input("Enter number 5: "))

# Calculate average
average = (num1 + num2 + num3 + num4 + num5) / 5

print(f"The average is: {average}")
```

```python
numbers = []

for i in range(5):
    num = float(input(f"Enter number {i+1}: "))
    numbers.append(num)

# Calculate average
average = sum(numbers) / len(numbers)

print(f"The numbers are: {numbers}")
print(f"The average is: {average}")
```

<div class="alert alert-block alert-success">
<b>Calculate percentage and grade from marks</b>
</div>

```python
subjects = int(input("Enter number of subjects: "))
marks = []
for i in range(subjects):
    mark = float(input(f"Enter marks for subject {i+1}: "))  # Fixed: added input()
    marks.append(mark)

total = sum(marks)
percentage = (total / (subjects * 100)) * 100

if percentage >= 90:
    grade = "A+"
elif percentage >= 80:
    grade = "A"
elif percentage >= 70:
    grade = "B"
elif percentage >= 60:
    grade = "C"
elif percentage >= 50:
    grade = "D"
elif percentage >= 40:
    grade = "E"
else:
    grade = "F"

print(f"Total Marks: {total}")
print(f"Percentage: {percentage:.2f}%")
print(f"Grade: {grade}")
```

</div>

```python
year = int(input("Enter a year:"))
if (year % 4 == 0) or (year % 100 == 0 and year % 4 == 0):
    print(f"{year} is a leap year")
else:
    print(f"{year} is not a leap year")
```

<div class="alert alert-block alert-success">
<b>Convert days into years, weeks, and days</b>
</div>

```python
total_days = int(input("Enter total days: "))

years = total_days // 365
remaining_days = total_days % 365
weeks = remaining_days // 7
remaining_days = remaining_days % 7

print(f"{total_days} days = {years} years, {weeks} weeks and {remaining_days} days")
```

<div class="alert alert-block alert-success">
<b>Calculate compound interest</b>
</div>

```python
principal = float(input("Enter a principal amount: "))
time = float(input("Enter a time: "))
rate = float(input("Enter a rate: "))

print(f"Compound Amount = {principal * ((1 + rate / 100)** time):.2f}")
print(f"Compound Interest = {(principal * ((1 + rate / 100)** time)) - principal}")
```
