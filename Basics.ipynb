```python
# Create a variable called name and assign your name. Print "Hello, [name]!"
name = input("Enter your name: ")
print(f"Hello, {name}")
```

```python
# Create two numbers a and b. Print their sum, difference, product, and quotient.
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
# I nested the input() inside the int() because input() returns strings, not numbers

print(f"The sum is {a+b}")
print(f"The difference is {a-b}")
print(f"The product is {a*b}")
print(f"The quotient is {a/b}")
```

```python
# Ask the user for their age using input(). Convert it to an integer and print how old they will be in 10 years.
user_age = int(input("Enter your age: "))
print(f"After 10 years, you will be {user_age + 10} years old")
```

```python
# Create variables for length and width. Calculate and print the area of a rectangle.
length = int(input("Enter length of the rectangle: "))
width = int(input("Enter width of the rectangle: "))

print(f"The area of the rectangle is {length * width}")
```

```python
# Ask the user for a temperature in Celsius. Convert it to Fahrenheit and print the result. (Formula: F = C * 9/5 + 32)
tempCelsius = float(input("Enter temperature in Celsius: "))
print(f"The temperature in Fahrenheit is {tempCelsius * 9/5 + 32}")
```

```python
# Swap two numbers without using a third variable
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))

print(f"Before swapping a = {a} and b = {b}")

a = a + b
b = a - b
a = a - b

print(f"After swapping a = {a} and b = {b}")
```
